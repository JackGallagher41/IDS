```python
# used for data analysis in Python

import pandas as pd
```

```python
# Pandas read_csv from url

url = "https://raw.githubusercontent.com/bforoura/IDS/main/alcohol_deaths.csv"

df = pd.read_csv(url, sep=",")
```

```python
df = df.append({'Country': 'India', 'Alcohol':2.53, 'Deaths': 792, 'Heart': 171, 'Liver': 'NA'}, ignore_index=True)

```

```python
# display the entire data frame

df
```

## **Problem 1**
Replace the negative numbers in Alcohol and Deaths with their absolute values:

```python
df['Alcohol'] = df['Alcohol'].abs();

```

```python
df['Deaths'] = df['Deaths'].abs()
```

## **Problem 2**
Impute all missing values using column means:

```python
df = df.fillna(df.mean())
```

## **Problem 3**
Discrete Deaths using your own judgement:

```python
df['Deaths'] = pd.cut(x=df['Deaths'], bins=[600, 700, 800, 900, 10000], labels = [0, 1, 2, 3])
```
