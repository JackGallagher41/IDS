## **1. Use the following dataset about adults: https://github.com/bforoura/IDS/blob/main/HW4/adult_training.csvLinks to an external site.**

```r
#Reads in the data file
library(readr)
adultTrainingDF <- read_csv("https://raw.githubusercontent.com/bforoura/IDS/main/HW4/adult_training.csv",show_col_types = FALSE)


#summarizes data and prints us first 10 rows
summary(adultTrainingDF)

head(adultTrainingDF, 10)
```

## **2. Create a contingency table of workclass and sex, and plot it.**

```r
#create the contingency table for workclass and sex
r_crosstab <- table(adultTrainingDF$sex,adultTrainingDF$workclass)

ftable(r_crosstab)
```

```r
#Create barplot to go with table use beside true, las and color to style graph and legend for key
colors <- c("navy", "orange")

barplot(r_crosstab, col = colors, beside = TRUE, las = 2)
legend("topright", legend = c("Female", "Male"), fill = colors)
```

Private Male tends to be the largest grouping while without pay and never worked for both male and female has little representation in the data frame. Even though male representation is nearly double in private the most representation female hold is in private as well.

## **3. Create a contingency table of sex and marital status, and plot it.**

```r
#create the contingency table for marital-status and sex
r_crosstabTwo <- table(adultTrainingDF$sex,adultTrainingDF$`marital-status`)

ftable(r_crosstabTwo)
```

```r
#Create barplot to go with table use beside true, las and color to style graph and legend for key
colors <- c("navy", "orange")

barplot(r_crosstabTwo, col = colors, beside = TRUE, las = 2)
legend("topright", legend = c("Female", "Male"), fill = colors)
```

When looking at the barplot married civ spouse has a huge spike in men while it is only the third highest marital status for women. The highest balanced plot is never married while lowest for both is married after spouse.

## **4. Create a new data frame, adultOver40, for those whose age > 40**

```r
#creates a df from the original with only paricipants older then 40
adultOver40DF <- adultTrainingDF[!(adultTrainingDF$age < 40), ]

adultOver40DF
```

## **5. Recreate the contingency table of sex and marital status for adultOver40**

```r
#Creates a new contingency table that is now referencing the new dataframe
r_crosstabThree <- table(adultOver40DF$sex,adultOver40DF$workclass)

ftable(r_crosstabThree)
```

```r
#Create barplot to go with table use beside true, las and color to style graph and legend for key
colors <- c("navy", "orange")

barplot(r_crosstabThree, col = colors, beside = TRUE, las = 2)
legend("topright", legend = c("Female", "Male"), fill = colors)
```

When comparing the over age 40 chart we can see it is vert similar even though it is a smaller scale. Another thing that is notable is that never worked is no long represented by anybody.

## **6. Determine whether any outliers exist for the education field**

```r
#First we start by scaling the education field
adultTrainingDF$scaled_education <- scale(x = adultTrainingDF$education)

summary(adultTrainingDF$scaled_education)
```

```r
#Check to see how many outliers there are through the three sigma rule and defining it
three_sigma_rule <- (adultTrainingDF$scaled_education < -3 |
                     adultTrainingDF$scaled_education > 3 )


#Checking count of outliers
cat ("There are", sum(three_sigma_rule), "outliers \n\n")
```

## **7. Standardize (zero mean) the age variable, and identify how many outliers there are. What is the most extreme outlier?**

```r
#First we start by scaling the age field
adultTrainingDF$scaled_age <- scale(x = adultTrainingDF$age)

summary(adultTrainingDF$scaled_age)
```

```r
#Check to see how many outliers there are through the three sigma rule and defining it
three_sigma_rule <- (adultTrainingDF$scaled_age < -3 |
                     adultTrainingDF$scaled_age > 3 )


#Checking count of outliers
cat ("There are", sum(three_sigma_rule), "outliers \n\n")
```

Using Max and Min z scores we can see that the min is not an outlier and 3.7512 is the Max z-score outlier. We can also determine that 90 is the age of the outlier because it is the max before scaled.

## **8. Age anomaly? Select only records with age  at least 80, and construct a histogram of age. Explain what you see.**

```r
#create a dataframe of anyone who is at least 80 years old
atLeast80DF <- adultTrainingDF[!(adultTrainingDF$age < 80), ]

atLeast80DF
```

```r
#form a histogram using column age for the new dataframe
hist(atLeast80DF$age,
       col = 'red',
       border = 'blue',
       main = paste("Age Histogram Over 80"),
       xlab = atLeast80DF$age,
       ylab = "Representation",
       breaks = 10,
       right = FALSE)
```

The most representation from ages over 80 is from 89-90 year olds
